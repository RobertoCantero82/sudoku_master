```python
from ultralytics import YOLO

# CONFIRMA que esta es la ruta a TU modelo entrenado, no a yolo11n.pt
modelo = YOLO(r'C:\Users\rober\Documents\Bootcamp\sudoku_master\app\modelos\modelo_yolo.pt')

# imprimo las clases que conoce el modelo: si está bien entrenado debería salir algo como 'sudoku'
print("Clases del modelo:", modelo.names)

ruta_imagen = r'C:\Users\rober\Downloads\Captura de pantalla 2026-06-17 152439.png'

# bajo el umbral de confianza para ver si detecta ALGO aunque sea con poca seguridad
results = modelo(ruta_imagen, conf=0.10)

# reviso cuántas cajas encontró y con qué confianza
cajas = results[0].boxes
print(f"\nNúmero de detecciones: {len(cajas)}")
for caja in cajas:
    clase = modelo.names[int(caja.cls.item())]
    print(f"  Clase: {clase} | Confianza: {caja.conf.item():.2f}")

results[0].show()
```
